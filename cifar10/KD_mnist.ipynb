```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import matplotlib.patches as patches
from tqdm import tqdm

import seaborn as sns
sns.set()
sns.set_style('white')
%matplotlib inline
```

```python
pickle_file = 'mnist_teacher_noise.pickle'
with open(pickle_file, "rb") as input_file:
    data = pickle.load(input_file, encoding="latin1")
```

```python
train_accuracy = data['train_accuracy']
test_accuracy = data['test_accuracy']
```

```python
noise_sigma_list = [0, .1, .2, .5, 1., 2., 5.]
```

```python
color_list = ['red', 'maroon', 'saddlebrown', 'orange', 'darkolivegreen', 
              'lime', 'navy', 'blue', 'purple', 'indigo']
fig = plt.figure(figsize=(10, 4))

for i in range(train_accuracy.shape[0]):
    plt.plot(noise_sigma_list, train_accuracy[i, :], linestyle='-', color = color_list[i])
    plt.plot(noise_sigma_list, test_accuracy[i, :], linestyle='--', color = color_list[i])
```
